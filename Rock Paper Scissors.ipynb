# Rock Paper Scissors (you vs computer) Let's see who's gonna win!

```python
import random
name=input("Enter you name to start the play with the system : ")
print("Hello "+name+"! Welcome to the play")
player_score=0
computer_score=0
while(True):
    if(player_score==5 or computer_score==5):
        break
    else:
        option=input("Choose between Rock, Paper, Scissors : ")
        choices = ["Rock", "Paper", "Scissors"]
        computer = random.choice(choices)
        if(computer=="Rock"):
            if(option=="Rock"):
                player_score+=0
                computer_score+=0
                print("Your choice : ",option)
                print("Computer choice : ",computer)
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
            elif(option=="Paper"):
                player_score+=1
                print("Your choice : ",option)
                print("Computer choice : ",computer)
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
            elif(option=="Scissors"):
                computer_score+=1
                print("Your choice : ",option)
                print("Computer choice : ",computer)
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
        elif(computer=="Paper"):
            if(option=="Rock"):
                computer_score+=1
                print("Your choice : ",option)
                print("Computer choice : ",computer)
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
            elif(option=="Paper"):
                player_score+=0
                computer_score+=0
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
            elif(option=="Scissors"):
                player_score+=1
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
        elif(computer=="Scissors"):
            if(option=="Rock"):
                player_score+=1
                print("Your choice : ",option)
                print("Computer choice : ",computer)
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
            elif(option=="Paper"):
                computer_score+=1
                print("Your choice : ",option)
                print("Computer choice : ",computer)
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
            elif(option=="Scissors"):
                player_score+=0
                computer_score+=0
                print("Your choice : ",option)
                print("Computer choice : ",computer)
                print("Your Score : ",player_score)
                print("Computer Score : ",computer_score)
if(player_score>computer_score):
    print("Wohoo! Master congratulations you won over the computer")
else:
    print("Next time better luck champ!...It's not that easy to compete with the Computer")
            
            
```
